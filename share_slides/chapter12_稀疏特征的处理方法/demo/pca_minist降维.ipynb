```python
digits = datasets.load_digits(n_class=10)
X = digits.data
y = digits.target
print (X.shape)
n_img_per_row = 20
img = np.zeros((10 * n_img_per_row, 10 * n_img_per_row))
for i in range(n_img_per_row):
    ix = 10 * i + 1
    for j in range(n_img_per_row):
        iy = 10 * j + 1
        img[ix:ix + 8, iy:iy + 8] = X[i * n_img_per_row + j].reshape((8, 8))
plt.imshow(img, cmap=plt.cm.binary)
plt.title('A selection from the 64-dimensional digits dataset')
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
X_new=pca.fit_transform(X)
```

```python
X_new.shape
```

```python
index_0=np.argwhere(y==0).reshape(-1,)
index_1=np.argwhere(y==1).reshape(-1,)
index_2=np.argwhere(y==2).reshape(-1,)
index_3=np.argwhere(y==3).reshape(-1,)
index_4=np.argwhere(y==4).reshape(-1,)
```

```python
for i in index_0:
    plt.plot(X_new[i][0],X_new[i][1],'.', color='b')
for i in index_1:
    plt.plot(X_new[i][0],X_new[i][1],'.', color='r')
for i in index_2:
    plt.plot(X_new[i][0],X_new[i][1],'.', color='g')
for i in index_3:
    plt.plot(X_new[i][0],X_new[i][1],'.', color='c')
for i in index_4:
    plt.plot(X_new[i][0],X_new[i][1],'.', color='y')

    
```
